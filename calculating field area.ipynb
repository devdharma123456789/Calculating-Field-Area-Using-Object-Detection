```python
import cv2
import numpy as np
cordinates = {}
# calculate the Lower and Upper range of HSV using Taskbar
lower_range6=np.array([122,203,69])
upper_range6=np.array([177,255,255])

lower_range1=np.array([0,186,222])
upper_range1=np.array([21,255,255])

lower_range=np.array([140,99,155])
upper_range=np.array([179,255,255])
 
lower_range2=np.array([22,73,210]) 
upper_range2=np.array([45,232,255])

lower_range3=np.array([72,208,132])
upper_range3=np.array([117,255,219])

# lower_range3=np.array([80,69,227])
# upper_range3=np.array([95,255,255])

lower_range4=np.array([94,177,229])
upper_range4=np.array([133,255,255])

lower_range5=np.array([129,66,102])
upper_range5=np.array([141,199,199])

# Load the image and Resize the image 
frame=cv2.imread("C:/Users/Dharma Dev/Desktop/Project/crop area/drone/003.jpg") # input as a images 
frame=cv2.resize(frame,(660,700))
# DarkRed color object detection and contour the object
def darkred(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range6,upper_range6)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for c in cnts:
        x=220
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c) 
# Find the centroid of Contour
            a1=int(x+x+w)//2
            b1=int(y+y+h)//2
            cv2.circle(img,(a1,b1),4,(0,0,0),-1)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
#             cv2.putText(frame, "center", (a1 - 20, b1 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.putText(frame,("k"),(a1+15,b1),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cv2.putText(frame,(str(a1)+", "+str(b1)),(a1-50,b1-20),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['k'] = (a1, b1)

# Red color object detection and contour the object
def red(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range,upper_range)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for c in cnts:
        x=220
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c) 
# Find the centroid of Contour
            x1=int(x+x+w)//2
            y1=int(y+y+h)//2
            cv2.circle(img,(x1,y1),4,(0,0,0),-1)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(frame,("R"),(x1+15,y1),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
#             cv2.putText(frame, "center", (x1 - 20, y1 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.putText(frame,(str(x1)+", "+str(y1)),(x1-50,y1-20),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['r'] = (x1, y1)
                  

# Orange color object dectection and contour the object
def orange(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range1,upper_range1)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    
    for c in cnts:
        x=220
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c)
# Find the centroid of Contour
            x2=int(x+x+w)//2
            y2=int(y+y+h)//2
            cv2.circle(img,(x2,y2),4,(0,0,0),-1)
#             cv2.putText(frame, "center", (x2 - 20, y2 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(frame,("O"),(x2-30,y2+15),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cv2.putText(frame,(str(x2)+", "+str(y2)),(x2-60,y2-20),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['o'] = (x2, y2)     
# Yellow color object dectection and contour the object
def yellow(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range2,upper_range2)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    for c in cnts:
        x=220
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c)
# Find the centroid of Contour
            x6=int(x+x+w)//2
            y6=int(y+y+h)//2
            cv2.circle(img,(x6,y6),4,(0,0,0),-1)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
#             cv2.putText(frame, "center", (x6 - 20, y6 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.putText(frame,("Y"),(x6+ 15,y6),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
#calculate the cordinate of centroid of yellow colors          
            cv2.putText(frame,(str(x6)+", "+str(y6)),(x6,y6-20),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['y'] = (x6, y6)
            
# Blue color object dectection and contour the object
def blue(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range3,upper_range3)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    b=cnts[0]
    for c in cnts:
        x=200
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c)
# Find the centroid of Contour
            x4=int(x+x+w)//2
            y4=int(y+y+h)//2
            cv2.circle(img,(x4,y4),4,(255,255,255),-1)
#             cv2.putText(frame, "center", (x4 - 20, y4 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(frame,("B"),(x4+15,y4),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
#calculate the cordinate of centroid of blue colors       
            cv2.putText(frame,(str(x4)+", "+str(y4)),(x4,y4-20),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['b'] = (x4, y4)
        
# Sky blue color object dectection and contour the object
def sky(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range4,upper_range4)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    s=cnts[0]
    for c in cnts:
        x=200
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c)
# Find the centroid of Contour         
            x3=int(x+x+w)//2
            y3=int(y+y+h)//2
            cv2.circle(img,(x3,y3),4,(0,0,0),-1)
#             cv2.putText(frame, "center", (x3 - 20, y3 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(frame,("S"),(x3+15,y3),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
#calculate the cordinate of centroid of sky colors
            cv2.putText(frame,(str(x3)+", "+str(y3)),(x3-30,y3+40),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['s'] = (x3, y3)        
# Voilet color object dectection and contour the object
def violet(img):
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    mask=cv2.inRange(hsv,lower_range5,upper_range5)
    _,mask1=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    cnts,_=cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    v=cnts[0]
    for c in cnts:
        x=200
        if cv2.contourArea(c)>x:
            x,y,w,h=cv2.boundingRect(c)
# Find the centroid of Contour
            x5=int(x+x+w)//2
            y5=int(y+y+h)//2
            cv2.circle(img,(x5,y5),4,(255,255,255),-1)
#             cv2.putText(frame, "center", (x5 - 20, y5 - 20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(frame,("V"),(x5+15,y5),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cv2.putText(frame,(str(x5)+", "+str(y5)),(x5,y5-20),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2)
            cordinates['v'] = (x5, y5)

            
while True:
    img1= cv2.imshow("frame",frame)
    darkred(frame)
    red(frame)
    orange(frame)
    yellow(frame)
    blue(frame)
    sky(frame)
    violet(frame)
    cv2.imwrite("test.jpg", frame)
    if cv2.waitKey(1) & 0xFF ==27:
        break  
        
cv2.destroyAllWindows()
```

```python
# output of this project 
import cv2

# Load the image
image = cv2.imread('C:/Users/Dharma Dev/Desktop/Project/crop area/drone/poll.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Calculate the area of the edges
largest_contour=max(contours,key=cv2.contourArea)
# for contour in contours:

area = cv2.contourArea(largest_contour)
area_m=round(area/RATIO_PIXEL_TO_SQUARE_M,2)

print(f"The area of polygon is {area_m} meter square")
cv2.drawContours(image,[largest_contour], -1, (0, 255, 0), 2) 
cv2.imshow('Edges', image)
cv2.waitKey(0)
cv2.imwrite("sarkar.jpg", image)
cv2.destroyAllWindows()
```
